## 데이터 세트 분리
### 예측 데이터 값을 기다린 후 다시 학습해야 되어 시간이 오래 걸리기 때문에 통상 필드에서는 분리 작업을 함.

훈련 세트(Train Set) : 테스트 세트(Test Set) = (80:20)

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# 훈련 80 : 테스트 20 으로 분리
# random_state : 훈련세트와 테스트 세트의 데이터를 랜덤하게... 기본값은 1, 사용 안할시 0
```

```python
X_train, len(X_train)
```

```python
y, len(y)
```

```python
y_train, len(y_train)
```

```python
y_test, len(y_test)
```

## 분리된 데이터 세트를 통한 모델링

훈련용 데이터 세트를 통해 학습을 진행하고 모델 생성 후
테스트 데이터 세트를 통해 예측 및 결과 평가를 진행

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train) # 훈련 세트로 학습한 모델
```

## 데이터 시각화 (훈련 세트)

```python
plt.scatter(X_train, y_train, color='green') # 산점도 그래프 (독립변수, 종속변수)
plt.plot(X_train, reg.predict(X_train), color='yellow') # 선 그래프 (예측 값)
plt.title('Score by Hours (Train Data)') # 제목
plt.xlabel('hours') # x 축 이름
plt.ylabel('score') # y 축 이름
plt.grid()
plt.show()
```

## 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test, y_test, color='green') # 산점도 그래프 (독립변수, 종속변수)
plt.plot(X_test, reg.predict(X_test), color='yellow') # 선 그래프 (예측 값)
plt.title('Score by Hours (Test Data)') # 제목
plt.xlabel('hours') # x 축 이름
plt.ylabel('score') # y 축 이름
plt.grid()
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

## 모델 평가

0 ~ 1 사이 (1로 가까워 질수록 모델의 성능이 높다)

```python
reg.score(X_test, y_test) # 테스트 세트를 통한 모델 평가
```

```python
reg.score(X_train, y_train) # 훈련 세트를 통한 모델 평가
```

## 경사 하강법 (Gradient Descent)

실제 값과 예측 값 차이(잔차)의 제곱의 합을 최소화 sum(y-yhate)^2

max_iter : 훈련 세트 반복 횟수 (Epoch 횟수)
eta0 : 학습률 (learning rate)

```python
from sklearn.linear_model import SGDRegressor # SGD : Stochastic Gradient Descent - 확률적 경사 하강법
sr = SGDRegressor()
sr.fit(X_train, y_train) # 모델 생성
```

```python
plt.scatter(X_train, y_train, color='green') # 산점도 그래프 (독립변수, 종속변수)
plt.plot(X_train, sr.predict(X_train), color='yellow') # 선 그래프 (예측 값)
plt.title('Score by Hours (Train Data - SGD)') # 제목
plt.xlabel('hours') # x 축 이름
plt.ylabel('score') # y 축 이름
plt.grid()
plt.show()
```

```python
sr.coef_
```

```python
sr.intercept_
```

```python
sr.score(X_test, y_test)
```

```python
sr.score(X_train, y_train)
```

## 지수 표기법
### 1e-3 : 0.001 (10의 -3승) 기본값
### 1e-4
### 1e+3 : 1000 (10의 3승)

```python
from sklearn.linear_model import SGDRegressor # SGD : Stochastic Gradient Descent - 확률적 경사 하강법
sr = SGDRegressor(eta0=1e-4, random_state=0, max_iter=1000, verbose=1)
sr.fit(X_train, y_train) # 모델 생성
```

```python
plt.scatter(X_train, y_train, color='green') # 산점도 그래프 (독립변수, 종속변수)
plt.plot(X_train, sr.predict(X_train), color='yellow') # 선 그래프 (예측 값)
plt.title('Score by Hours (Train Data - SGD)') # 제목
plt.xlabel('hours') # x 축 이름
plt.ylabel('score') # y 축 이름
plt.grid()
plt.show()
```
