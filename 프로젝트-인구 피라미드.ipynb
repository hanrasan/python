# 인구 피라미드

연령별 인구 현황 - (https://jumin.mois.go.kr/)

```python
import pandas as pd
file_name = '201303_201303_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(5)
```

## 남자 데이터 정의

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 69,830 -> 69830 (정수형)
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(5)
```

```python
df_w.columns = df_m.columns # column 명이 다르게 생성되어 동일하게 적용함
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w.iloc[0]
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기 지정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, df_m.iloc[0] // 1000) # 단위를 1000명으로 바꿈
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
```

```python
plt.figure(figsize=(10,7))
plt1=plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # -df_m에서 '-'의 의미는 왼쪽 위치에
plt1=plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2013년 대한민국 인구 피라미드')
plt.savefig('2013년_인구피라미드.png')
plt.show()
```
