# 그룹화

동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해서 사용

```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col='학생번호')
df
```

```python
# 그룹화하기 위해 groupby 함수 사용
# - 그룹화가 먼저하고 데이터확인은 후에 해야함

df.groupby('학교') 
# 그룹화되었다는 알림이 뜸.(오류는 아님) 
#- <pandas.core.groupby.generic.DataFrameGroupBy object at 0x000002A24D97B9A0>
```

```python
# get_group : 그룹화된것을 가져온다.
df.groupby('학교').get_group('북산고')
```

```python
df.groupby('학교').get_group('능남고')
```

```python
# 그룹화된 학교별로 평균보기

# 계산 가능한 데이터들의 평균값을 출력
df.groupby('학교').mean()

# GPT 로 오류확인 결과(참고)
# -> df.groupby('학교').mean(numeric_only=True) 
#   numeric_only 인자를 True로 설정
# -> df.select_dtypes(include='number').groupby('학교').mean() 
#   숫자 데이터만 선택해서 평균값을 계산
```

```python
df.groupby('학교').size() # 각 그룹의 크기
```

```python
# 학교로 그룹화를 한 뒤에 능남고에 해당하는 데이터의 수를 출력
df.groupby('학교').size()['능남고']
```

```python
# 학교로 그룹화를 한 뒤에 키의 평균 데이터
df.groupby('학교')['키'].mean()
```

```python
# 학교로 그룹화를 한 뒤에 국,영,수 의 평균 데이터
df.groupby('학교')[['국어', '영어', '수학']].mean()
```

```python
# 학년 Column 추가

df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
# 학교별 학년별 평균데이터
df.groupby(['학교', '학년']).mean()

# GPT 오류검색 결과(참고)
# ->  df.groupby(['학교', '학년']).mean(numeric_only=True) 
#     numeric_only 인자를 True로 설정
# -> df.select_dtypes(include='number').groupby(['학교', '학년']).mean() 
#    숫자 데이터만 선택해서 평균값을 계산
```

```python
df.groupby(['학년']).mean()

# GPT 오류검색 결과(참고)
# -> df.groupby(['학년']).mean(numeric_only=True) 
#    numeric_only 인자를 True로 설정
# -> df.select_dtypes(include='number').groupby(['학년']).mean() # 숫자 데이터만 선택해서 평균값을 계산
```

```python
df.groupby('학년').mean().sort_values('키', ascending=False)

# GPT 오류검색 결과(참고)
# -> df.groupby('학년').mean(numeric_only=True).sort_values('키', ascending=False) 
#    numeric_only 인자를 True로 설정
# -> df.select_dtypes(include='number').groupby('학년').mean().sort_values('키', ascending=False) 
#    숫자 데이터만 선택해서 평균값을 계산 후 '키' 컬럼을 기준으로 내림차순으로 정렬
```

```python
df.groupby(['학교', '학년']).sum()

# GPT 오류검색 결과(참고)
# -> df.groupby(['학교', '학년']).sum(numeric_only=True) # numeric_only 인자를 True로 설정
# -> df.select_dtypes(include='number').groupby(['학교', '학년']).sum() 
#    숫자 데이터만 선택해서 합계를 계산
```

```python
# 학교로 그룹화를 한 뒤, 이름, SW 특기의 데이터의 수를 가져옴
# 데이터의 개수를 확인할때는 count() 함수 사용
df.groupby('학교')[['이름', 'SW특기']].count()
```

```python
# 학교로 그룹화를 한 뒤에 학년별로 그룹화를 하여 학생 수를 가져옴
school = df.groupby('학교')
school['학년'].value_counts()
```

```python
# 학교로 그룹화를 한 뒤 학년별로 그룹화를 하여 '북산고'의 학생 수를 가져옴
school['학년'].value_counts().loc['북산고']
```

```python
school['학년'].value_counts().loc['능남고']
```

```python
# 학교 / 학년으로 그룹화해서 북산고 학생들의 수의 퍼센트
school['학년'].value_counts(normalize=True).loc['북산고']
```
