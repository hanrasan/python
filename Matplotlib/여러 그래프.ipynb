# 여러 그래프

## 하나의 영역에 여러 개의 그래프를 함께 넣기

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기 지정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10)) # 2 X 2 에 해당하는 그래프들을 생성
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 실행
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].legend() # 범례
axs[0, 0].set(xlabel='이름', ylabel='점수') # X, Y 축 이름
axs[0, 0].set_facecolor('yellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학점수')
axs[0, 1].plot(df['이름'], df['영어'], label='영어점수')

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], label='사회점수')
```
