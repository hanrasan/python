```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 데이터 생성
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# 3차원 스캐터 플롯 생성
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c=z, cmap='Blues')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
x = np.linspace(-5, 5, 100)
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y, label='sine curve', color='blue', linestyle='dashed', linewidth=2)
plt.title('Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

# 그래프 출력
plt.show()
```

```python
import numpy as np

# x, y 값의 범위 설정
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

# x, y 값으로 2차원 배열 생성
X, Y = np.meshgrid(x, y)

# z 값 계산
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3차원 그래프 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# x, y, z 값으로 그래프 그리기
ax.plot_surface(X, Y, Z)

# 축 라벨 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 그래프 제목 설정
plt.title('3D Surface plot')

# 그래프 출력
plt.show()
```
