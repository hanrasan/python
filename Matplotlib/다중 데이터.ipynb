#  다중 데이터

## 디증 그래프

데이터가 다중으로 있는 경우

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기 지정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용시, 마이너스 글자가 깨지는 현상을 해결

x = [1, 2, 3]
y = [2, 4, 8]

plt.plot(x, y)
```

## COVID-19 백신 종류별 접종 연구

```python
days = [1, 2, 3] # 1일, 2일, 3일
az = [2, 4, 8] # [ 단위 : 만명 ] 1일부터 3일까지 az 접종한 인구 데이터
pfizer = [5, 1, 3]
moderna = [1, 2, 3]

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, moderna, label='moderna', marker='s', ls='-.')

plt.legend(ncol=3)
```
