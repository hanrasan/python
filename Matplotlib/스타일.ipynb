# 스타일

그래프를 꾸밀 때 사용하는 것

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기 지정
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth=5)
```

## 마커(Marker) 선(Line)

x, y 데이터의 값이 만나는 위치 

```python
plt.plot(x, y, marker='o', linewidth=5, linestyle='-', color='g', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

```python
plt.plot(x, y, marker='o', linewidth=5, linestyle='None')
```

## 포멧

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'yv-.')
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'gx')
```

```python
plt.plot(x, y, color='g', marker = 'x', linestyle="None")
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls='-', alpha=0.2) # alpha 투명도 ( 0 ~ 1 사이의 값을 가진다. )
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(4,3), dpi=100) # DPI - Dots Per Inch
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='red')
plt.plot(x, y)
```
